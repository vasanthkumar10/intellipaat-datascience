```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# print(plt.__version__)
print(np.__version__)
print(pd.__version__)
```

```python
# Basic graph
# plt.plot([1, 2, 3], [1, 2, 3])

x = [2, 4, 6, 8, 10]
y = [1, 2, 3, 4, 5]


# plt.plot(x, y)
# plt.plot(x, y, label='2x', color='red', linestyle='--', linewidth=2, marker='.', markersize=10, markeredgecolor='black')
plt.plot(x, y, label='2x', color='red', linestyle=':', linewidth=2, marker='.', markersize=10, markeredgecolor='black')
# {'-', '--', '-.', ':', ''}

# shorthand
# '[color],[marker][line]'
# plt.plot(x, y, 'r.:', label='2x')

# plt.legend()

# plt.title('Test graph', fontdict={'fontname': 'consolas'})
plt.title('Test graph', fontdict={'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel('X axis', fontdict={'fontname': 'consolas'})
plt.ylabel('Y axis', fontdict={'fontname': 'consolas'})

# plt.xticks(range(0, 10, 1))
# plt.yticks(range(0, 10, 1))

# plt.xticks([2, 4, 6, 8, 1])
plt.xticks([2, 4, 6, 8, 10])
# plt.yticks([1, 2, 3, 4, 5])

# resize graph
# plt.figure(figsize=(15, 3), dpi=300)
# plt.figure(figsize=(15, 3))

# line2
x2 = np.arange(0, 5, 1)
plt.plot(x2, x2**3, label='x^3')
plt.plot(x2[:2], x2[:2]**3, label='x^3')
plt.legend()

# save
# plt.savefig('testgraph.png')
plt.savefig('testgraph1.png', dpi=100)


plt.show()
```

```python
# Bar chart

x = ['A', 'B', 'C']
y = [2, 5, 3]

plt.figure(figsize=(4, 5))

# plt.bar(x, y)
bar_chart = plt.bar(x, y)
bar_chart[0].set_hatch('/')
bar_chart[1].set_hatch('.')
bar_chart[2].set_hatch('*')

plt.title('bar chart')

plt.show()
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
x = range(32)
y = df['hp']


# plt.figure(figsize=(2, 5))
# plt.title('Hp')
plt.plot(x, y, label='hp')
plt.show()
```

```python
x2 = range(32)
y2 = df['disp']

plt.figure(figsize=(4, 5))
plt.title('disp')

plt.plot(x2, y2)
plt.show()
```

```python
hp = df['hp']
disp = df['disp']

plt.plot(hp, disp, color='red', label='power')
plt.xlabel('hp')
plt.ylabel('disp')
plt.legend()
plt.show()
```

```python
y1 = df['hp']
y2 = df['disp']

x = range(32)
plt.plot(x, y1, label='hp')
plt.plot(x, y2, label='disp')
plt.legend()

plt.show()
```

```python
# bar chart
y1 = df['hp']
y2 = df['disp']

x = range(32)
plt.bar(x, y1, label='hp')
plt.bar(x, y2, label='disp')
plt.legend()

plt.show()
```

```python
# bar chart
y1 = df['gear']
y2 = df['disp']

x = range(32)
plt.bar(x, y1, label='gear')
plt.bar(x, y2, label='disp')
plt.legend()

plt.show()
```
